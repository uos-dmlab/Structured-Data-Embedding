```python
#딕셔너리 형태로 변환
tab_dic = {}
for t in Table_column_pair_list:
    tab_col1, tab_col2 = t.split(" @ ")
    tab1,col1 = tab_col1.split("^")
    tab2,col2 = tab_col2.split("^")
    if tab_col1 not in tab_dic:
        tab_dic[tab_col1] = [tab_col2]
    else:
        tab_dic[tab_col1].append(tab_col2)
```

```python
def containment(arr1,arr2):
    #모든 값 소문자로 변경
    larr1 = set([str(i).lower() for i in arr1])
    larr2 = set([str(i).lower() for i in arr2])
    
    threshold = 0.9
    denominator = min(len(larr1), len(larr2))
    numerator = len(larr1.intersection(larr2))
    
    if (numerator / denominator) > threshold:
        return True
    else:
        return False
```

```python
#N-Join-Pair의 입력을 만들때 a-b, a-c 테이블쌍을 활용해 a-b-c를 만든다. 이 때, b-c의 containment 값이 일정 수준 이상임을 확인할 필요가 있다. 이를 확인하는 코드가 아래에 있다.
join_list3_1 = []
for j in tqdm(join_list2):
    tab_col1, tab_col2 = j.split(" @ ")

    if tab_col1 in tab_dic.keys():
        for v in tab_dic[tab_col1]:
            _a = set(pd.read_csv(f"./데이터셋_20240520/{tab_col2.split('^')[0]}")[f'{tab_col2.split("^")[1]}'])
            _b = set(pd.read_csv(f"./데이터셋_20240520/{v.split('^')[0]}")[f'{v.split("^")[1]}'])
            if v != tab_col2 and containment(_a,_b):
                
                join_list3_1.append(j + " @ " + v)

```

```python
#조인 가능한 테이블쌍이 a-b , b-c와 같이 되어있는 경우 a-b-c를 만든다. 이 때, a-c의 containment 값이 일정 수준 이상임을 확인할 필요가 있다. 이를 확인하는 코드가 아래에 있다.
join_list3_2 = []
for j in tqdm(join_list2):
    tab_col1, tab_col2 = j.split(" @ ")
    
    if tab_col2 in tab_dic.keys():
        for v in tab_dic[tab_col2]:
            _a = set(pd.read_csv(f"./데이터셋_20240520/{tab_col1.split('^')[0]}")[f'{tab_col1.split("^")[1]}'])
            _b = set(pd.read_csv(f"./데이터셋_20240520/{v.split('^')[0]}")[f'{v.split("^")[1]}'])
            if v != tab_col1 and (j + " @ " + v) not in join_list3_1 and containment(_a,_b):
                join_list3_2.append(j + " @ " + v)
```
